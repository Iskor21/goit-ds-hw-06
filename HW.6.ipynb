```python
# напишіть функцію гіпотези лінійної регресії у векторному вигляді;
import numpy as np

def hypothesis(X, theta):
    """
    X: матриця ознак (m x n)
    theta: вектор параметрів (n x 1)
    return: вектор передбачень (m x 1)
    """
    return np.dot(X, theta)

# Приклад:
# Матриця ознак (3 приклади, 2 ознаки + 1 для константи)
X = np.array([[1, 2, 3],
              [1, 4, 5],
              [1, 6, 7]])

# Вектор параметрів (theta0, theta1, theta2)
theta = np.array([0.5, 1.0, -0.5])

# Передбачення
predictions = hypothesis(X, theta)
print(predictions)
```

```python
# створіть функцію для обчислення функції втрат у векторному вигляді;
def loss_function(X, y, theta):
    m = len(y)
    predictions = X @ theta          # h_theta(x) = Xθ
    errors = predictions - y         # різниця між прогнозом і реальністю
    cost = (1 / (2 * m)) * (errors.T @ errors)
    return cost.item()               # повертаємо скаляр

# Матриця ознак (3 приклади, 2 ознаки + bias)
X = np.array([[1, 2, 3],
              [1, 4, 5],
              [1, 6, 7]])

# Вектор правильних значень
y = np.array([1, 2, 3])

# Вектор параметрів
theta = np.array([0.5, 1.0, -0.5])

# Обчислення функції втрат
J = loss_function(X, y, theta)
print("Функція втрат J(theta):", J)
```

```python
# реалізуйте один крок градієнтного спуску;
def gradient_descent_step(X, y, theta, alpha):
    """
    Виконує один крок градієнтного спуску для лінійної регресії.
    
    X: матриця ознак (m x n)
    y: вектор правильних значень (m x 1)
    theta: вектор параметрів (n x 1)
    alpha: швидкість навчання
    return: оновлений вектор параметрів theta
    """
    m = len(y)
    predictions = X @ theta              # h_theta(x) = Xθ
    errors = predictions - y             # різниця між прогнозом і реальністю
    gradient = (1/m) * (X.T @ errors)    # векторний градієнт
    theta = theta - alpha * gradient     # оновлення параметрів
    return theta

# Матриця ознак (3 приклади, 2 ознаки + bias)
X = np.array([[1, 2, 3],
              [1, 4, 5],
              [1, 6, 7]])

# Вектор правильних значень
y = np.array([1, 2, 3])

# Початкові параметри
theta = np.array([0.5, 1.0, -0.5])

# Швидкість навчання
alpha = 0.01

# Один крок градієнтного спуску
theta_updated = gradient_descent_step(X, y, theta, alpha)
print("Оновлені параметри θ:", theta_updated)
```

```python
# знайдіть найкращі параметри для датасету
def hypothesis(X, w):
    return X @ w

def loss_function(X, y, w):
    m = len(y)
    errors = hypothesis(X, w) - y
    return (1/(2*m)) * (errors.T @ errors)

def gradient_descent_step(X, y, w, alpha):
    m = len(y)
    errors = hypothesis(X, w) - y
    gradient = (1/m) * (X.T @ errors)
    return w - alpha * gradient

def train_linear_regression(X, y, alpha=0.01, iterations=1000):
    # Ініціалізація ваг нулями
    w = np.zeros(X.shape[1])
    for i in range(iterations):
        w = gradient_descent_step(X, y, w, alpha)
        if i % 100 == 0:  # кожні 100 ітерацій показуємо втрату
            print(f"Iteration {i}, Loss: {loss_function(X, y, w)}")
    return w
    
# Дані: площа, ванні, спальні
X = np.array([[1, 120, 2, 3],   # 1 = bias term
              [1, 150, 2, 4],
              [1, 200, 3, 4],
              [1, 80, 1, 2]])

# Ціни (y)
y = np.array([100000, 150000, 200000, 80000])

# Навчання
best_w = train_linear_regression(X, y, alpha=0.0000001, iterations=1000)
print("Найкращі параметри w:", best_w)
```

```python
# знайдіть ці ж параметри за допомогою аналітичного рішення;
def normal_equation(X, y):
    """
    Аналітичне рішення для лінійної регресії.
    
    X: матриця ознак (m x n)
    y: вектор правильних значень (m x 1)
    return: оптимальні параметри w
    """
    XTX = X.T @ X
    XTy = X.T @ y
    w = np.linalg.inv(XTX) @ XTy
    return w

# Дані: площа, ванні, спальні (з bias term = 1)
X = np.array([[1, 120, 2, 3],
              [1, 150, 2, 4],
              [1, 200, 3, 4],
              [1, 80, 1, 2]])

# Ціни
y = np.array([100000, 150000, 200000, 80000])

# Аналітичне рішення
best_w = normal_equation(X, y)
print("Найкращі параметри w:", best_w)
```

```python
# для перевірки спрогнозованих значень, використайте LinearRegression з бібліотеки scikit-learn та порівняйте результати.
from sklearn.linear_model import LinearRegression

# Дані: площа, ванні, спальні
X = np.array([[120, 2, 3],
              [150, 2, 4],
              [200, 3, 4],
              [80, 1, 2]])

# Ціни
y = np.array([100000, 150000, 200000, 80000])

# --- Аналітичне рішення (нормальне рівняння) ---
# додаємо bias term (стовпець з 1)
X_bias = np.c_[np.ones(X.shape[0]), X]
w_analytical = np.linalg.inv(X_bias.T @ X_bias) @ (X_bias.T @ y)
print("Аналітичні параметри w:", w_analytical)

# --- Перевірка через scikit-learn ---
model = LinearRegression(fit_intercept=True)
model.fit(X, y)

print("Intercept (w0):", model.intercept_)
print("Coefficients (w1, w2, w3):", model.coef_)

# --- Порівняння ---
print("\nПорівняння результатів:")
print("Аналітичне w0:", w_analytical[0], " | sklearn intercept:", model.intercept_)
print("Аналітичні коефіцієнти:", w_analytical[1:], " | sklearn coef:", model.coef_)
```
